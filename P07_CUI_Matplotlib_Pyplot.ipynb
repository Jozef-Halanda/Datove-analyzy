```python
import matplotlib.pyplot as plt
import numpy as np

skupiny = ["1. skupina", "2. skupina", "3. skupina"]
hodnoty = [1,10,100]

plt.figure(figsize=(11.5,3.5))
# default 6.4, 4.8 
plt.subplot(1,4,1)
# Cislo riadku, Cislo stlpca, index
# nrows, ncols, index
# default 111(1,1,1)
plt.bar(skupiny, hodnoty)
plt.subplot(1,4,2)
plt.plot(skupiny, hodnoty)
plt.subplot(1,4,3)
plt.scatter(skupiny, hodnoty)
plt.subplot(1,4,4)
plt.plot(skupiny, hodnoty)


plt.show(skupiny, hodnoty)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Stredna hodnota/median
# Sigma smerodajna/standardna odchylka +- autobus
mi, sigma = 100, 20 # mi(mu)
x = mi + sigma*np.random.randn(10000)
n, bins, patches = plt.hist(x, 20, density = 1, facecolor="m", alpha = 0.8)# density-hustota,facecolor-farba,alpha-priehladnost

plt.xlabel("IQ")
plt.ylabel("Pravdepodobnosti")
plt.title("Histogram IQ")
plt.text(50, 0.015, r"$\mu\ = 100$")# r-formatovat, zacat aj ukoncit $ - r"$\beta\$"
plt.text(50, 0.0125, r"$\sigma\ = 20$")
plt.grid()
plt.show()
```

```python
import tkinter
import random

canvas = tkinter.Canvas()
canvas.pack()

for i in range(10):
    x = random.randint(10,100)
    y = random.randint(10,100)
    sirka, vyska = 200, 60
    canvas.create_rectangle(x, y, x + sirka, y + vyska, fill = "salmon")
    
tkinter.mainloop()
```
