# 02. Dátové typy (typy premenných)

```python
# str (string) - reťazec
meno_klient = "Adam";
print("meno_klient ma typ premennej: ", type(meno_klient));
    
meno_klient = input("Ako sa mas? -> ");
print("Ahoj", meno_klient);

```

```python
# int (integer) - celé čísla - Z
vek_klient = 23;
print(type(vek_klient));
print(vek_klient);
```

```python
# float - reálne čísla - R - desatinná bodka
mzda_klient = 2389.63;
print(type(mzda_klient));
print(mzda_klient);
```

```python
# bool/boolean - Logicke hodnoty - True (pravda), False (nepravda)
status_klient_zenaty = True;
print(type(status_klient_zenaty));
print(status_klient_zenaty);
```

```python
registracia_klient = None
print(type(registracia_klient));
print(registracia_klient);
```

```python
# complex - Komplexné číslo
# 5 +2i
a = 3 + 3j;
print(type(a));
```

```python
import datetime
aktualny_datum_cas = datetime.datetime.now();
print("Aktuálny dátum a čas je:",aktualny_datum_cas);
aktualny_rok = aktualny_datum_cas.year;
print("Aktuálny rok je:", aktualny_rok);
```

```python
# bytes - (bajty)
a = b"Ahoj";
print(type(a));
print(a);
```

```python
# Memoryview (adresy v pameti) 0-9A-F, 10->A, 11->B, 15->F
b = memoryview(bytes(2));
print(type(b));
print(b);
```

# 03 Pretypovanie (casting)

```python
vyplata = 1956.23
print("výplata ma hodnotu: ", vyplata, "je to typ: ", type(vyplata))
vyplata_cele_cisla = int(vyplata);
print("výplata cele čísla ma hodnotu: ", vyplata_cele_cisla, "je to typ: ", type(vyplata_cele_cisla));

vyplata_retazec = str(vyplata);
print("výplata retazec ma hodnotu: ", vyplata_retazec, "je to typ: ", type(vyplata_retazec));

# 0, None,"" --> false
vyplata_bool = bool(vyplata);
print("výplata bool ma hodnotu: ", vyplata_bool, "je to typ: ", type(vyplata_bool));

```
